```
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```
img = cv2.imread('20210426.JPG')
img = img[:,:,::-1] # BGR -> RGB
img.shape
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
img_gray.shape
plt.imshow(img_gray,cmap='gray')
```

```
ret, img_th = cv2.threshold(img_gray, 120, 180, cv2.THRESH_BINARY)
plt.imshow(img_th, cmap='gray')
```

```
img_blurred = cv2.GaussianBlur(img_gray, (91, 91), 0)
plt.imshow(img_blurred, cmap='gray')
```

```
kernel = np.array([[-1,-1,-1], 
          [-1, 9,-1],
          [-1,-1,-1]])
img_sharpened = cv2.filter2D(img_gray, -1, kernel)
plt.imshow(img_sharpened, cmap='gray')
```

```
img_canny = cv2.Canny(img_gray, 40, 50)
plt.imshow(img_canny, cmap='gray')
```
